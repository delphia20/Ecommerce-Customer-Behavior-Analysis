```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\merli\OneDrive\Desktop\E-commerce Customer Behavior - Sheet1.csv")
df.head()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
print(df.columns)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['satisfaction_level'] = df['satisfaction_level'].fillna(
    df['satisfaction_level'].mode()[0]
)
```

```python
print(df.isnull().sum())
```

```python
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
print(df['membership_type'].unique())
```

```python
frequency_mapping = {
    'Gold': 'Weekly',
    'Silver': 'Bi-weekly',
    'Bronze': 'Monthly'
}

df['purchase_frequency'] = df['membership_type'].map(frequency_mapping)
```
